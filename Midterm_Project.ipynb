```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("ScreenTime vs MentalWellness.csv")

df["screen_time_hours"] = pd.to_numeric(df["screen_time_hours"], errors="coerce")
df["mental_wellness_index_0_100"] = pd.to_numeric(df["mental_wellness_index_0_100"], errors="coerce")

df = df.dropna(subset=["screen_time_hours", "mental_wellness_index_0_100"])

df["screen_category"] = pd.cut(
    df["screen_time_hours"],
    bins=[0, 2, 4, 6, 10],
    labels=["0-2 hrs", "2-4 hrs", "4-6 hrs", "6+ hrs"]
)

df["screen_category"] = pd.Categorical(
    df["screen_category"],
    categories=["0-2 hrs", "2-4 hrs", "4-6 hrs", "6+ hrs"],
    ordered=True
)

df = df.sort_values("screen_category")

fig = px.box(
    df,
    x="screen_category",
    y="mental_wellness_index_0_100",
    title="Impact of Screen Time on Mental Wellness",
    labels={
        "screen_category": "Screen Time (Hours)",
        "mental_wellness_index_0_100": "Mental Wellness Score"
    }
)

fig.update_layout(
    template="plotly_white",
    title_x=0.5
)

fig.update_traces(marker_color="#636EFA")

fig.add_annotation(
    text="Higher screen time is associated with lower mental wellness",
    x=0.5,
    y=1.05,
    xref="paper",
    yref="paper",
    showarrow=False,
    font=dict(size=12)
)

fig.show()
```
